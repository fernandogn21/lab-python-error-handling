   - If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, display an error message and ask them to re-enter the product name. *Hint: you will need to pass inventory as a parameter*
   - Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.

````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            quantity = input(f"Enter the quantity of {product}s available: ")
            if quantity.isdigit():
                quantity = int(quantity)
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            else:
                print("Invalid input. Please enter a valid quantity.")
    return inventory

def get_customer_orders(inventory):
    valid_num_orders = False
    while not valid_num_orders:
        num_orders = input("Enter the number of customer orders: ")
        if num_orders.isdigit():
            num_orders = int(num_orders)
            if num_orders >= 0:
                valid_num_orders = True
            else:
                print("Number of orders cannot be negative. Please enter a valid number.")
        else:
            print("Invalid input. Please enter a valid number.")
    
    customer_orders = []
    for _ in range(num_orders):
        valid_product = False
        while not valid_product:
            product = input("Enter the name of a product that a customer wants to order: ")
            if product in inventory and inventory[product] > 0:
                customer_orders.append(product)
                valid_product = True
            else:
                print("Invalid product name or out of stock. Please enter a valid product name.")
    
    return customer_orders

def calculate_total_price(customer_orders):
    prices = []
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            price = input(f"Enter the price of {product}: ")
            if price.replace('.', '', 1).isdigit():  # Allows for decimal points
                price = float(price)
                if price >= 0:
                    prices.append(price)
                    valid_price = True
                else:
                    print("Price cannot be negative. Please enter a valid price.")
            else:
                print("Invalid input. Please enter a valid price.")
    total_price = sum(prices)
    return total_price

def update_inventory(inventory, customer_orders):
    for order in customer_orders:
        if order in inventory:
            inventory[order] -= 1
    inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
    return inventory

def run_order_management():
    # Initial Inventory
    products = ["t-shirt", "mug", "hat", "book", "keychain"]
    
    # Step 1: Initialize Inventory
    inventory = initialize_inventory(products)
    
    # Step 2: Get Customer Orders
    customer_orders = get_customer_orders(inventory)
    
    # Step 3: Update Inventory
    inventory = update_inventory(inventory, customer_orders)
    
    # Step 4: Calculate Total Price
    total_price = calculate_total_price(customer_orders)
    
    # Print Customer Orders
    print("\nCustomer Orders:")
    for order in customer_orders:
        print(order)
    
    # Print Order Statistics
    print("\nOrder Statistics:")
    total_products_ordered = len(customer_orders)
    percentage_unique_products_ordered = (len(set(customer_orders)) / len(products)) * 100
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Unique Products Ordered: {percentage_unique_products_ordered:.1f}%")
    
    # Print Updated Inventory
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
    
    # Print Total Price
    print(f"Total Price: {total_price}")

if __name__ == "__main__":
    run_order_management()
````
